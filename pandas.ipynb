```
import pandas as pd
url="https://raw.githubusercontent.com/ryali93/UNMSM_programacion/master/data/earthquakes.csv"
a=pd.read_csv(url)
data=a[(a["Latitude"]<0) & (a["Magnitude"]>7)]
len(data)
```

```
from geopandas import GeoDataFrame
from shapely.geometry import Point
```

```
#GeoDataFrame()
# df # Data frame
crs={'init': 'epsg:4326'}
geometry=[Point(x) for x in zip(a.Longitude, a.Latitude)]

```

```
a= GeoDataFrame (a, crs=crs, geometry=geometry)
a.head()
```

```
b=a[(a["Latitude"]<0) & (a["Magnitude"]>7)]
b.plot()
```

```
import folium
```

```
geojson = b.to_json()
```

```
mapa = folium.Map([-12,-72], zoom_start=8,tiles="OpenStreetMap")
points = folium.features.GeoJson(geojson)
mapa.add_children(points)
mapa
```
