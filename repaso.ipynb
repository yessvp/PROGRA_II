```
def triang (a,b,c):
  per = a+b+c
  s = per/2
  ar = (s)*(s-a)*(s-b)*(s-c)**0.5
  return per, ar

triang(4,3,7)
```

```
def factorial(x):
  fact = 1
  for m in range (1, x+1):
    fact = fact*m
  return fact
    
```
